```python
import pandas as pd
```

```python
students_df = pd.read_csv("files/students_record.csv")

print("DataFrame shape (rows, columns):", students_df.shape)
print("\nColumn names (in the form of a list):", students_df.columns.tolist())

print("\nData types:")
print(students_df.dtypes)

# Summary statistics
print("\nStatistical summary of dataset:")
print(students_df.describe())
```

```python
# Accessing columns
print("\nLast Names:")
print(students_df['last_name'])
```

```python
# Filtering - In this case we're filtering for students with math score of 90 and above
high_scores = students_df[students_df["math_score"] >= 90]
print(high_scores)
```

```python
# Conditional selection
new_df = students_df[["student_id", "first_name", "last_name", "gender", "age", "attendance_rate"]]
print(new_df)
```
